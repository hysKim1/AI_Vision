## Convolutional Neural Network
### Why Deep Learning
전통적 머신러닝, 딥러닝 차이

- ML: 직접 feature extraction하기 때문에 모델 분리되어 있음 
    - 머신러닝은 데이터를 기반으로 학습을 통해서  규칙을 스스로 찾음
        - 규칙에는 구멍이 존재
    - **데이터**가 많을수록 구멍이 최소로 있어야함
- DL: end-to-end model: 학습을 통해 feature extraction 자동화 하여 모델까지 함께 함
    - 연산량이 많지만 2012년 Imagenet 이후로 각광
    

### FCNN
- Fully Connected Neural Network
    - 학습을 통해서 실제값과 예측값 차이(loss)를 최소화 하는 **weight, bias**를 조절하여 찾음 
- 모든 weight 계산(이미지의 모든 픽셀의 연관관계를 한번에 모두 연산) > 비효율성 
- **위치에 민감**
    > -  mnist 숫자분류 : 데이터가 배경없는 객체의 이미지가 정중앙에 맞춰줘 있어 flatten하면 모든 값들이 비슷한 특성 패턴으로 나타남.
      - 이미지에서 특성이 일부만 나오거나, 다른 방향인 경우 인식 불가됨 
-  동일 feature(pixel)가 다른 위치에 있으면 똑같은 feature로 input되지 않고 hidden-Layer에서 동일한 feature을 중복해서 뽑아냄
    - 이차원 이미지는 이차원적인 특성을 갖는데 ANN으로는 힘듦
## CNN

### 공통된 Architecture
1. 학습데이터 입력 
2. Feed forward
3. output - 손실함수loss functon 계산하여 최소화를 목적
4. 최적화를 이용해 가중치, bias 업데이트

### CNN Architecture
- CNN(Convolutional Neural Network)은 인접 이미지와의 특징을 효과적으로 인식(공간 정보)하고 강조하는 방식으로 복수의 필터로 이미지의 **특징을 추출(feature extraction)** 과 이미지를 분류기(classifier)로 구성
  


1. Feature Extraction
    - 특징 추출 영역은 Filter를 사용하여 공유 파라미터 수를 최소화하면서 이미지의 특징을 찾는 Convolution 레이어와 **특징을 강화하고 모으는 Pooling 레이어**로 구성
    - CNN은 Filter의 크기, Stride, Padding과 Pooling 크기로 출력 데이터 크기를 조절하고, 필터의 개수로 출력 데이터 의 채널을 결정
    - 각 레이어의 입출력 데이터의 형상 유지
    - 필터를 공유 파라미터로 사용하기 때문에,  FCNN 와 비교하여 더 작은 학습 파라미터로 더 높은 인식률
    
  A. convolution
    - 이미지와 가중치들의 집합체인 다양한 필터와은 컨볼루션 연산
  (cross-correlation연산) 통해서 각 필터별 **특징 추출(feature map)**
     - Input * Filter +bias = feature map
        - 필터를 이동 간격(stride)로 sliding하면서 element-wise product 을해서 합함
        
  B. activation function
    - relu: $max(0,x)$
    
  C. pooling
    - 데이터를 최대/최소/평균값으로 압축하여 연산량을 줄임 
    - feature map을 압축한 pooling 값: 클수록 필터의 특징을 많이 포함하기 때문에 특징 추출
    - 예. 숫자 11은 세로 특징이 더 많아 세로 필터가 유용

    -  **합성곱  -> feeature map -> activation function -> activation map**
    - Feature Map: 합성곱 계산으로 만들어진 행렬
    - Activation Map: Feature Map 행렬에 활성 함수를 적용한 결과 


2. FCNN
    - 학습을 통해서 실제값과 예측값 차이(loss)를 최소화 하는 **kernel**을 조절하여 찾음


    


### CNN 특징
1. ### locally-connected
    - 특정 픽셀에 대해서 **거리가 가까운 pixel은 연관성이 있어서 계산**
    - 거리가 먼 pixel은 연관성이 없는 것으로 보고서 계산하지 않는 방식
2. ### Convolutional Filter
    - cross- correlation 연산
     > ### convolution
        -  영상처리에서 convolution 을 filter로 사용 
    -  이미지가 locally-conntected한 이미지 데이터라는 것을 가정해야 나타나는 개념
    - (현재위치의 1pixel과 그 가로세로의 크기만큼의 주변pixel의 연관관계를 계산 - 가로x세로의 Kernel=Filter를 sliding 하면서 연산)
    - input의 위치와 관계없이, **shared weight(특정 영역 추출)**를 sliding window하고, 그것을 연산하여, 똑 같은 feature라면 **위치와 상관없이 해당feature를 뽑아낼 수 있음**

                          batch matrix product
    - 6개의 Convolutional Filter: 서로 다른 랜덤 초기화 필터
    - output feature: 각각 연산해서 각각 6개의 output을 만들고 그 output은 새로운 feature로서 6개가 산출
        - input_filter_size와 convolutional filter의 filter_size는 같아야 함
        - output_filter_size와 convolutional Layer의 filter개수는 같아야 함
    * Convolutional Layer:여러개의 Convolution Filter를 묶어놓은 것
        - 하나의 Filter당 하나의 feature를 뽑아내며  여러 개의 다양한 feature를 뽑아내기 위해서는 여러개의 Filter를 사용
            - 데이터가 많이 필요하므로 최소의 특징으로 분류를 잘하는것을 찾음
            
       -  channel별 여러개의 다른 커널 사용
            > black-box model : 사람이 이해할 수 없는 특징들을 학습을 통해 추출
       -  Layer는 앞쪽일수록 저차원의 linear한 feature를 뽑아내고, 점점 더 추상화된 고차원 feature (**hierarchy** 특성) 추출

  

3. ### Padding 
    - Sub sampling
    - 입력 데이터 주변을 특정 값으로 둘러싸서 채움
    
        - zero-padding : 가장자리를 “0”값으로 둘러싸는 특징
    - Convolution 레이어의 **출력 데이터의 사이즈를 조절**하여  결과가 **작아지는 것을 막아서 특징이 유실되는 것을 막음**
        - 출력 = (입력+2패딩-필터)/스트라이드 +1
        - 가장자리는 연산 횟수가 적어서 동일한 연산을 해주도록 패딩하여 이미지 중앙에 오버피팅 방지
    - 원본 데이타에 값을 넣어서 원래의 특징을 희석 시켜 버리고, 이것을 기반으로 ML 모델이 트레이닝 값에만 정확하게 맞아 들어가는 **오버피팅 방지**

    ### 커널 크기
    - VGG 실험에서는 3x3 , 5x5 size
    > NFL theorem


4. ### Pooling
    - **데이터를 공간적 특성을 유지하며 작은 사이즈로 압축**
        -  필요없는 feature를 버려서, 남은 메모리는 다음 Conv Layer를 쌓는데 쓰는, **연산을 효율적**으로 만듦 
        - 연산에 들어가는 컴퓨팅 리소스가 적음
        -  소실이 발생하기 때문에, 오버피팅을 방지
        - Pooling 레이어를 통과하면 행렬의 크기 감소 & 채널 수 무변
    - 특징을 나타내는 픽셀은 그렇게 많기 때문에 중요한 pixel만 선별해서 연산
    - **invariant** 특성 
       - 고유의 변하지 않는 특성이 있어서 어느정도 왜곡에 강해짐
        - NFL  -> stride 로 해결
    - 학습대상 파라미터가 없음
    
    ### Max Pooling
    - 특징을 나타내는 픽셀중에서 가장 큰 값 선택
    - 실용적인 접근에 있어서 많이 안쓰려고 하는 추세
    ### stride
    - pooling 안 쓰고 stride 크기를 키워서 poolingr과 같은 효과

## 흑백 이미지

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data() #holdout

X_train = X_train.reshape((-1, 28, 28, 1))  # 6만개의 28x28 흑백 이미지
X_test = X_test.reshape((-1, 28, 28, 1))    # 1만개의 28x28 흑백 이미지

X_train, X_test = X_train / 255.0, X_test/ 255.0  #  0~1 사이로 정규화
```

##### Functional API

```python
input_ = tf.keras.Input(shape=(28,28,1))   
conv1=tf.keras.layers.Conv2D(32, 3 ,activation='relu')(input_)
mpool1=tf.keras.layers.MaxPool2D(2)(conv1)
conv2=tf.keras.layers.Conv2D(64, 3, activation='relu')(mpool1)
mpool1=tf.keras.layers.MaxPool2D(2)(conv2)


flatten =tf.keras.layers.Flatten()(mpool1)
dense1 =tf.keras.layers.Dense(64, activation ='relu')(flatten)
dense2 =tf.keras.layers.Dense(10, activation ='softmax')(dense1)

model= tf.keras.Model(inputs=input_, outputs=dense2)
```

```python
model.summary()
```

##### Sequential API

```python
#Sequential API 이해용도로 사용
model=tf.keras.models.Sequential([
    tf.keras.Input(shape=(28,28,1)),
    tf.keras.layers.Conv2D(32,(3,3), activation='relu'),
    tf.keras.layers.MaxPool2D((2, 2)),
    tf.keras.layers.Conv2D(64,(3,3), activation='relu'),
    tf.keras.layers.MaxPool2D((2, 2)),    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(64, activation ='relu'),
    tf.keras.layers.Dense(10, activation ='softmax')
])
```

```python
#  상동
model = tf.keras.models.Sequential()
model.add(tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))
model.add(tf.keras.layers.Conv2D(64, (3, 3), activation='relu'))
model.add(tf.keras.layers.MaxPooling2D((2, 2)))

model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

- 연산량이 많아서 FCNN에 비해서 느림 
- 데이터 자체가 전처리 잘 되어 있어서 정확도가 99%로 높음

```python
#fit: gradient tape이용 low-level로 커스터마이징
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

model.fit(X_train, y_train, epochs=5)
```

```python
model.summary()
```
